```python
# Install required library (run once)
# !pip install mlxtend

# Import libraries
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# -------------------------------
# 1. Load Dataset
# -------------------------------
df = pd.read_excel("/content/Online Retail.xlsx")

# Preview
print(df.head())

# -------------------------------
# 2. Data Cleaning
# -------------------------------

# Remove missing descriptions
df = df.dropna(subset=['Description'])

# Remove cancelled invoices (InvoiceNo starting with 'C')
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]

# Remove negative or zero quantities
df = df[df['Quantity'] > 0]

# -------------------------------
# 3. Create Basket (One-Hot Encoding)
# -------------------------------

basket = (df
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum()
          .unstack()
          .fillna(0))

# Convert quantities to 0/1
basket = basket.applymap(lambda x: 1 if x > 0 else 0)

print(basket.head())

# -------------------------------
# 4. Apply Apriori Algorithm
# -------------------------------

frequent_items = apriori(basket,
                         min_support=0.02,  # adjust as needed
                         use_colnames=True)

print(frequent_items.head())

# -------------------------------
# 5. Generate Association Rules
# -------------------------------

rules = association_rules(frequent_items,
                         metric="lift",
                         min_threshold=1)

# Sort rules by lift
rules = rules.sort_values(by='lift', ascending=False)

print(rules.head())

# -------------------------------
# 6. Filter Strong Rules
# -------------------------------

strong_rules = rules[(rules['confidence'] > 0.5) & (rules['lift'] > 1)]

print(strong_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```
